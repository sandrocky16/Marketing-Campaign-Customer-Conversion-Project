```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
new=pd.DataFrame(Sub_No['Age_Category'].value_counts())
new.reset_index(inplace = True)
new
```

## 📊 EDA Insights Summary (Non-Subscribed Customers)
1. The majority of non-converting customers belong to the 30–45 age group, with the 31–35 segment contributing the highest share.

2. Within this segment, customers in management roles form a significant portion of non-subscribers, indicating that higher professional status does not necessarily translate to higher conversion.

3. A large proportion of these customers are single, suggesting that marital status may influence financial decision-making and subscription behavior.

4. Customers with “unknown” previous campaign outcomes show low conversion rates, indicating that cold leads are less responsive to direct marketing efforts.

5. Despite high contact frequency, especially during peak months like May, August, and November, conversion remains low, showing that more outreach does not guarantee better results.

6. The analysis reveals a low ROI customer segment (mid-age, management, single, unknown history) where marketing efforts are being inefficiently allocated.

7. Overall, customer profiling and segmentation play a more critical role in conversion than simply increasing campaign intensity.
